# Insights

## Total number of fire in each state 

```python
# State and Number

sub_data = df2.loc[:,["state","number"]]
sum_ = sub_data.groupby("state", as_index = False).sum()
```

```python
sum_["number"].max()
```

```python
sum_ = sum_.round(0)
sum_
```

```python
plt.figure(figsize = (14,8))
plot_data = sns.barplot(data = sum_, x = "number", y = "state", orient = "h", hue = "state")
plot_data.set_xlim(left = 0, right = 108000)
for i in plot_data.containers:
    plot_data.bar_label(i)
plt.title("Total number of fire in each state (1998- 2017)")
plt.show()
```

## Alagoas

```python
alagoas = df2[df2["state"] == "Alagoas"]
alagoas = alagoas.loc[:,["month","number"]]
alagoas = alagoas.groupby("month", as_index = False, sort = False).sum()
alagoas
```

```python
plt.figure(figsize = (15,7))
sns.lineplot(data = alagoas, x = "month", y = "number")
for i in range(len(alagoas)):
    plt.annotate(
        text=f"({alagoas['month'][i]}, {alagoas['number'][i]})",  # Text to display
        xy=(alagoas['month'][i], alagoas['number'][i]),           # Point to annotate
        xytext=(5, 5),                             # Offset for text
        textcoords='offset points',               # Coordinate system for offset
        fontsize=9,                                # Font size
        color='black'                              # Text color
    )
plt.title("ALAGOAS")
plt.show()
```

## Amapa

```python
amapa = df2[df2["state"] == "Amapa"]
amapa = amapa.loc[:,["month","number"]]
amapa = amapa.groupby("month", sort = False, as_index = False).sum()
amapa
```

```python
plt.figure(figsize = (15, 7))
sns.lineplot(data = amapa, x = "month", y = "number", color = "red")

for i in range(len(amapa)):
    plt.annotate(
        text = f"({amapa["number"][i]})",
        xy = (amapa["month"][i], amapa["number"][i]),
        xytext = (5,5),
        textcoords = "offset points"
    )
    
plt.title("AMAPA")
plt.show()
```
